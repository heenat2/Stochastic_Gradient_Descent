```python
import numpy as np
global a,b

# Initialize the hyperplane parameters a and b to 0
a = np.zeros((6,1))
b = 0.0

# Load data array
x = np.load('nparray.npy')
```

```python
def gamma_i(x_i,a_vec,b_sca):
    '''Function accepts a feature vector x_i, a and b and returns the value of a linear equation used for class prediction'''
    return np.dot(np.transpose(a_vec),x_i) + b_sca
```

```python
'''This code snippet trains a SVM using stochastic gradient descent to optimize a hinge loss function. The regularization 
constant is set to 0.01. The training proceeds in 100 epochs of 500 steps each. The dataset is divided into a 90% training set
and a 10% test set. Batch size is set to 1.'''
np.random.shuffle(x)
part_index = int(.9 * x.shape[0])
training = x[:part_index,:]
test = x[part_index:,:]
training_x = training[:,0:6]
training_y = training[:,6]
test_x = test[:,0:6]
test_y = test[:,6]
    
m,n,k,lambda_ = 1,50,500,0.01
batch_size = 1
print('lambda is',lambda_)
for epoch in range(1,101):
    #step size function of epoch number and constants m and n
    eta = m/(0.01*epoch + n)
    #list of accuracies on heldout data every 30 steps
    accuracy_list = []
    np.random.shuffle(training)
    #hold out 50 samples for evaluation of classifier
    heldout = training[:50,:]
    heldout_x = heldout[:,0:6]
    heldout_y = heldout[:,6]
    train = training[50:,:]
    train_x = train[:,0:6]
    train_y = train[:,6]
    for step in range(0,k):
        #select a train record at random and revise a and b
        train_k = train[np.random.choice(train.shape[0], batch_size, replace=True)]
        x_k, y_k = train_k[:,0:6].reshape(6,1), train_k[:,6]
        if y_k * gamma_i(x_k,a,b) >= 1:
            a_new = a - eta*lambda_*a
            a = a_new
        else:
            a_new = a - eta*(lambda_*a - y_k*x_k)
            b_new = b + eta*y_k
            a, b = a_new, b_new
        #compute accuracy of classifier on heldout data
        if step % 30==0:
            predicted_y = np.array([gamma_i(x,a,b) for x in heldout_x])
            accuracy_arr = (predicted_y * heldout_y.reshape(heldout_y.shape[0],1)) > 0
            accuracy_list.append(np.sum(accuracy_arr) / accuracy_arr.size)
    print('epoch is ',epoch, ' accuracy on heldout data: ',sum(accuracy_list)/len(accuracy_list))
print('done')
```

```python
# compute accuracy on 10% test data
print('a= ',a)
print('b= ',b)
predicted_y = np.array([gamma_i(x,a,b) for x in test_x])
accuracy_arr = (predicted_y * test_y.reshape(test_y.shape[0],1)) > 0
print('accuracy on test data is ',np.sum(accuracy_arr) / accuracy_arr.size)
```
